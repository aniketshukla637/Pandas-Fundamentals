```python
import pandas as pd
# Pandas library ko import kiya hai.
```

```python
# Titanic dataset load karna
# Hum ek publicly available CSV file se data load karenge.
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("Titanic dataset successfully load ho gaya hai.")
```

```python
# DataFrame ki pehli 5 rows dekhna
# df.head() se data ka structure aur pehli kuch entries dikhti hain.
print("\nDataFrame ki pehli 5 rows (df.head()):")
print(df.head())
```

```python
# DataFrame ki basic information check karna
# df.info() columns, non-null values aur data types batata hai.
print("\nDataFrame ki jaankari (df.info()):")
df.info()
```

```python
# DataFrame ke descriptive statistics dekhna
# df.describe() numerical columns ka summary deta hai (mean, std, min, max, etc.).
print("\nDescriptive Statistics (df.describe()):")
print(df.describe())
```

```python
# Missing values check karna
# df.isnull().sum() har column mein kitni missing values hain, ye batata hai.
print("\nMissing values har column mein:")
print(df.isnull().sum())
```

```python
# Ek specific column ke unique values aur unki count dekhna
# Jaise, 'Sex' column mein kitne male aur female hain.
print("\n'Sex' column ke unique values aur count:")
print(df['Sex'].value_counts())
```

```python
# Condition ke hisab se rows filter karna
# Sirf 'Female' passengers ka data dekhna.
female_passengers = df[df['Sex'] == 'female']
print("\nFemale passengers ka data (pehli 5 rows):")
print(female_passengers.head())
```

```python
# Naya column add karna
# Example ke liye, ek 'FamilySize' column banate hain 'SibSp' aur 'Parch' ko jodkar.
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print("\n'FamilySize' column add karne ke baad DataFrame (pehli 5 rows):")
print(df[['SibSp', 'Parch', 'FamilySize']].head())
```

```python
# Data Grouping karna
# 'Sex' aur 'Survived' ke hisab se count dekhna.
survived_by_sex = df.groupby('Sex')['Survived'].value_counts()
print("\nSurvived by Sex (Groupby):")
print(survived_by_sex)
```

```python
# DataFrame ko sort karna
# 'Age' column ke according data ko ascending order mein sort karna.
sorted_df = df.sort_values(by='Age', ascending=True)
print("\nDataFrame sorted by 'Age' (first 5 rows):")
print(sorted_df[['Name', 'Age', 'Survived']].head())
```

```python
# Columns drop karna
# Kuch unnecessary columns ko remove karna (e.g., 'Cabin').
df_dropped = df.drop(columns=['Cabin', 'PassengerId'])
print("\nDataFrame after dropping 'Cabin' and 'PassengerId' columns (first 5 rows):")
print(df_dropped.head())
```

```python
# Conditional selection with multiple conditions
# Un passengers ko dekhna jo 'female' hain aur 'Survived' hain.
survived_females = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print("\nSurvived Females (pehli 5 rows):")
print(survived_females.head())
```
