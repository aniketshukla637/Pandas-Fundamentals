```python
# Import the pandas library, commonly aliased as pd
import pandas as pd
# Define a list of numbers named 'x'
x=[1,2,3,4,5]
# Create a pandas Series named 'y' from the list 'x'
y=pd.Series(x)
# Print the pandas Series 'y'
print(y)
```

```python
import numpy as mp # Import the NumPy library (aliased as mp)
import pandas as pd # Import the Pandas library (aliased as pd)
marks=[67,34,56,78,89] # List of marks
subject=['maths','english','science','social','fghj'] # List of subjects
y=pd.Series(marks) # Create a Series 'y' from marks with default integer index
print(y) # Print Series 'y'
x=pd.Series(marks,index=subject) # Create a Series 'x' from marks with subject names as index
print(x) # Print Series 'x'
```

```python
import numpy as np
import pandas as pd
Marks = {
    'maths':167,
    'english':57,
    'science':34,
    'hindi':100,
}
print(marks)
marks_series=pd.Series(Marks,name='Marks')
print(marks_series)
print(marks_series.size)
print(marks_series.dtype)
print(marks_series.index)
print(marks_series.values)
print(marks_series.name)
print(marks_series.head(2))
print(marks_series.tail)
#sare sath mai print karake likhenge to sabko output ayega verna first and last ka sirf
```
